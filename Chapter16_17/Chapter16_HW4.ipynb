```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
TaxTable = pd.read_excel('Chapter16_17.xlsx', sheet_name=1) 
TaxTable = TaxTable.dropna()
TaxTable
```

```python
ax1 = TaxTable.plot.scatter(x="Highest Corporate Tax Rate",y="Unemployment Rate (July 2014)",c='DarkBlue', figsize=(10,10)) 
```

```python
r, pvalue = stats.spearmanr(TaxTable["Highest Corporate Tax Rate"], TaxTable["Unemployment Rate (July 2014)"])
print("r:", r)
print("pvalue:", pvalue)
```
