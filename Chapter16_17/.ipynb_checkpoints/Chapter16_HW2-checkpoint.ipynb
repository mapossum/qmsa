```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
GUNSandMurders = pd.read_excel('Chapter16_17.xlsx', sheet_name=2) 
GUNSandMurders
```

```python
ax1 = GUNSandMurders.plot.scatter(x="Percent Gun Ownershp",y="Gun Murders Per 100,000",c='DarkBlue', figsize=(10,10)) 
```

```python
r, pvalue = stats.pearsonr(GUNSandMurders["Percent Gun Ownershp"], GUNSandMurders["Gun Murders Per 100,000"])
print("r:", r)
print("pvalue:", pvalue)
```
