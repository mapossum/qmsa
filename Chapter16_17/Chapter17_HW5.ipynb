```python
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
DOTEMP = pd.read_excel('Chapter16_17.xlsx', sheet_name=4) 
DOTEMP
```

```python
ax1 = DOTEMP.plot.scatter(x="Temperature (degrees celsius)",y="DO (mg/L)",c='DarkBlue', figsize=(10,10))  
x = DOTEMP["Temperature (degrees celsius)"]
y = DOTEMP["DO (mg/L)"]

regressStats = stats.linregress(x, y)
slope = regressStats.slope
intercept = regressStats.intercept
print("slope:", slope)
print("intercept:", intercept)
plt.plot(x, slope * x + intercept, color="red")
```

```python
rsq = regressStats.rvalue**2
print("R-squared: %f" % rsq)
```

```python
DOTEMP['modeled_DO'] = DOTEMP["Temperature (degrees celsius)"] * slope + intercept
DOTEMP['residual'] = (DOTEMP['modeled_DO'] - DOTEMP['DO (mg/L)'])**2
DOTEMP
```

```python
SE = (DOTEMP['residual'].sum()/ (DOTEMP['residual'].count() - 2))**0.5
print("standard error", SE)
```

```python
Fstat = rsq * (DOTEMP['residual'].count() - 2) / (1 - rsq)
"test statistic", Fstat
```

```python
pvalueSLR = 1 - stats.f.cdf(Fstat, 1, DOTEMP['residual'].count() - 2)
"pvalue", pvalueSLR
```
