```python
# Read housingPrices Table
import pandas as pd
housingPrices = pd.read_excel('Chapter10.xlsx', sheet_name=0) 
housingPrices
```

```python
from scipy import stats
import numpy as np

#This one requires two samples.  It is the Two Sample Difference in Means Test (Independent)

a = housingPrices["Sale Price for New Homes in Lamar County, 2005 "]
b = housingPrices["Sale Price for New Homes in Lamar County, 2010 "]
print("2005 mean and standard deviation", a.mean(), a.std())
print("2010 mean and standard deviation", b.mean(), b.std())
stats.ttest_ind(a,b, equal_var=True)
```
