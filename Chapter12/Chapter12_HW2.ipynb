```python
import pandas as pd
import numpy as np
from arcgis.features import GeoAccessor, GeoSeriesAccessor
import matplotlib.pyplot as plt
from scipy import stats

sdf = pd.DataFrame.spatial.from_featureclass("Chapter12.gdb\GenerateTessellation_Spatial")
sdf = sdf[sdf['Join_Count'] < 12]
sdf
```

```python
tornadoCount = sdf.groupby(['Join_Count']).count()
tornadoCount
```

```python
tornadoCount = tornadoCount.reset_index()
tornadoCount['Observed'] = tornadoCount['OBJECTID']
tornadoCount = tornadoCount.drop(columns=['OBJECTID', 'SHAPE'])
lam = sdf['Join_Count'].mean()
tornadoCount['Expected'] = stats.poisson.pmf(tornadoCount['Join_Count'],lam) * sdf['Join_Count'].count()
tornadoCount
```

```python
plotTornadoCount = tornadoCount.set_index("Join_Count")[['Observed', 'Expected']]
ax = plotTornadoCount.plot.bar(rot=0)
```

```python
plotTornadoCount["ChiSqContrib"] = (plotTornadoCount['Observed'] - plotTornadoCount['Expected'])**2 / plotTornadoCount['Expected']
plotTornadoCount
```

```python
stats.chisquare(tornadoCount['Observed'], f_exp=tornadoCount['Expected'])
```
