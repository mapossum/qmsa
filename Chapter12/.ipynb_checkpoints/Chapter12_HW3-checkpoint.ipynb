```python
import pandas as pd
import numpy as np
from arcgis.features import GeoAccessor, GeoSeriesAccessor
import matplotlib.pyplot as plt
from scipy import stats

PollenData = pd.read_excel('Chapter12.xlsx', sheet_name=0) 
PollenData = PollenData.set_index('Species')
PollenData
```

```python
PollenData.transpose().plot.bar(stacked=True, figsize=(10,7))
```

```python
PollenData.plot.bar(stacked=True, figsize=(10,7))
```

```python
output = stats.chi2_contingency(PollenData, correction=False)
print("The value of the test statistic is: {0}".format(output[0]))
print("The p-value is: {0}".format(output[1]))
print("The Expected values are: \n{0}".format(output[3]))
```

```python
expectedValues = pd.DataFrame(output[3])
expectedValues.transpose().plot.bar(stacked=True, figsize=(10,7))
```

```python
expectedValues.plot.bar(stacked=True, figsize=(10,7))
```

```python
overallData = PollenData.reset_index().join(expectedValues)
overallData
```
