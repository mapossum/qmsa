```python
# Read Cucumber Data
import pandas as pd
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('TestData.xlsx', sheet_name=1) 
data
```

```python
ax1 = data.plot.scatter(x="Precipitation (mm)",y="Cucumbers Yield (Kg)",c='DarkBlue', figsize=(10,10)) 
```

```python
r, pvalue = stats.pearsonr(data["Precipitation (mm)"], data["Cucumbers Yield (Kg)"])
print("r:", r)
print("pvalue:", pvalue)
```

```python
ax1 = data.plot.scatter(x="Precipitation (mm)",y="Cucumbers Yield (Kg)",c='DarkBlue', figsize=(10,10)) 
x = data["Precipitation (mm)"]
y = data["Cucumbers Yield (Kg)"]

regressStats = stats.linregress(x, y)
m = regressStats.slope
b = regressStats.intercept
plt.plot(x, m * x + b, color="red")
```

```python
print("R-squared: %f" % regressStats.rvalue**2)
```
