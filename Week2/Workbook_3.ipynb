### Question 4:

```python
# First lets cut out all the data we don't need and put into a new dataframe
USsm = USDATA[["State", "Percent Urban 2010"]]
USsm
```

```python
# Here are the Unweighted values (Think about how you can get variance and coef. of var from these values)
USsm['Percent Urban 2010'].agg(["mean", "std"])
```

```python
#Divide into 5 equal bins using cut function, and create a field to hold the bin number
#Please use Percent Urban 2010, because the others actually don't have enough data in each class
USsm["Class"] = pd.cut(x=USsm["Percent Urban 2010"], bins=5, labels=[1,2,3,4,5])
USsm.sort_values(by=["Percent Urban 2010"])
```

```python
# Then group by one of the zones and calculate the aggregate values
# Use this table to calcuate the weighted mean, variance and Coeffient of Variation 
# Doing this by hand may be helpful.  There are only 5 values so it won't be hard.
# Look to the book for the formulas.
USgrouped = USsm.groupby(["Class"]).agg(["mean", "std", "count"])
USgrouped
```
