```python
import pandas as pd
import numpy as np
from arcgis.features import GeoAccessor, GeoSeriesAccessor
import matplotlib.pyplot as plt
from scipy import stats

sdf = pd.DataFrame.spatial.from_featureclass("Chapter13_14.gdb\CA_Points")
sdf 
```

```python
x, y = sdf["elevation_CA"], sdf["elevation_CA_nearby"]
regressStats = stats.linregress(x, y)
rsq = regressStats.rvalue**2
print("R-squared: %f" % rsq)
ax1 = sdf.plot.scatter(x="elevation_CA",y="elevation_CA_nearby",c='DarkBlue', figsize=(10,10)) 
```
